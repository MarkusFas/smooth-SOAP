```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def func(x, mu, sig):
    return np.exp(-((x-mu)**2)/(2*sig*sig))/(np.sqrt(2*np.pi)*sig)
```

```python
# define a RV 
mu = 0
sig = 3 
N = 100000
X = np.random.normal(mu, sig, N)

x = np.linspace(np.min(X), np.max(X), 100)
y = func(x, mu, sig)
hist, bin_edges = np.histogram(X, 25)
bin_centers = 0.5 * (bin_edges[:-1] + bin_edges[1:])
dens = hist/np.sum(hist)
print(np.sum(hist))

plt.bar(bin_centers, dens)
plt.plot(x,y, color='C1')
```
